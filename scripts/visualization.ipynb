<h3>Data Cleaning </h3>

```python
#loading the dataset
import pandas as pd

df = pd.read_csv('../datasets/spotify_data.csv')

print("dataset preview: ")
print(df.head(5))
```

```python
#dataset info and shape
print("dataset info: ")
df.info()
print("dataset shape: ")
df.shape

df.describe(include='all')
```

```python
#checking for duplicate rows
duplicates = df.duplicated().sum()
print(f"Total exact duplicate rows: {duplicates}")
```

```python
import numpy as np

df1 = df.replace('7', np.nan)
df1.info()
```

```python
#finding missing values in each column
missing_values = df1.isnull().sum()
print("Missing values in each column:")
print(missing_values)


missing_prcent = (df1.isnull().sum() / len(df1)) * 100
print("Percentage of missing values in each column:")
print(missing_prcent)
```

```python
#visualizing missing values
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Filter and Sort
missing_to_plot = missing_prcent[missing_prcent > 0].sort_values(ascending=False)

# 2. Setup the Plot
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# 3. Create Barplot
ax = sns.barplot(
    x=missing_to_plot.index, 
    y=missing_to_plot.values, 
    hue=missing_to_plot.index, 
    palette='flare',
    legend=False
)

# 4. CRITICAL FIX: Set Y-axis to Log Scale
# This makes 0.03% and 0.06% visible next to 39%
ax.set_yscale("log")

# 5. Add Labels
plt.title('Missing Values (Log Scale to see small percentages)', fontsize=15)
plt.ylabel('Percentage Missing (Log Scale %)', fontsize=12)
plt.xlabel('Columns', fontsize=12)

# 6. Add the exact value labels on top so there is no confusion
for p in ax.patches:
    ax.annotate(f'{p.get_height():.4f}%', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

<h3>Handling missing values </h3>

artist_name =  3 missing (MCAR) //drpoping rows

album_name = 6missing (MCAR) //filling them with unknown

artist_genres = 39% missing (MNAR)// unknown


```python
#dropping missing rows for artist name 
df1 = df1.dropna(subset=['artist_name'])
#filling missing values for album name and artist genres
df1['album_name'] = df1['album_name'].fillna('Unknown Album')
df1['artist_genres'] = df1['artist_genres'].fillna('Uncategorized')

df1.info()
print(df1.isnull().sum())
```

```python
# Converting data types
df1['album_release_date'] = pd.to_datetime(df1['album_release_date'])
df1['album_type'] = df1['album_type'].astype('category')

# Extracting year from release date : feature engineering
df1['release_year'] = df1['album_release_date'].dt.year

df1.info()
```

```python
#saving cleaned data to a new csv file
df1.to_csv('cleaded_data.csv', index = False)
```

<h2>DATA VISUALIZATION </h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Ensure release_year exists
if 'release_year' not in df1.columns:
	df1['album_release_date'] = pd.to_datetime(df1['album_release_date'], errors='coerce')
	df1['release_year'] = df1['album_release_date'].dt.year

#frequency plots 

#popularity distribution
plt.figure(figsize=(10,6))
sns.histplot(df1['track_popularity'], bins=30, kde=True, color="#93016aa1")
plt.title('Distribution of Song Popularity')
plt.xlabel('Track Popularity')
plt.ylabel('Frequency')
plt.show()
plt.savefig('popularity_distribution.png')

# Number of songs released each year
plt.figure(figsize=(12,6))
year_order = sorted(df1['release_year'].dropna().unique())
sns.countplot(x='release_year', data=df1, palette='coolwarm', order=year_order)
plt.title('Number of Songs Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
plt.savefig('songs_per_year.png')

# Top 10 Artists by Number of Songs
plt.figure(figsize=(12,6))
top_artists = df1['artist_name'].value_counts().head(10)
sns.barplot(x=top_artists.index, y=top_artists.values, palette='magma')
plt.title('Top 10 Artists by Number of Songs')
plt.xlabel('Artist Name')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
plt.savefig('top_artists_songs.png')

# Top 10 Artists by Total Followers
top_artists = df1[['artist_name', 'artist_followers']].drop_duplicates().nlargest(10, 'artist_followers')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_artists, x='artist_followers', y='artist_name', palette='flare')
plt.title('Top Artists by Total Followers')
plt.xlabel('Followers (Millions)')
plt.ylabel('Artist Name')
plt.show()
plt.savefig('top_artists_followers.png')

# Explicit vs. Non-Explicit Track Popularity
plt.figure(figsize=(8, 6))
sns.boxplot(x='explicit', y='track_popularity', data=df1, palette='magma')
plt.title('Popularity: Explicit vs. Non-Explicit Tracks')
plt.xlabel('Is Explicit?')
plt.ylabel('Track Popularity')
plt.show()
plt.savefig('explicit_vs_popularity.png')
```

```python


# 2. Proportion of Explicit Tracks (Pie Chart)
proportion_explicit = df1['explicit'].value_counts(normalize=True) * 100
plt.figure(figsize=(6,6))
plt.pie(proportion_explicit, labels=['Non-Explicit', 'Explicit'], autopct='%1.1f%%', 
        colors=["#f2a1f78c","#6b0167a3"], startangle=90, explode=(0, 0.1)) # Explode pops out the 'Explicit' slice
plt.title('Proportion of Explicit vs. Non-Explicit Tracks')
plt.show()

# 3. Proportion of Album Types
proportion_album_types = df1['album_type'].value_counts(normalize=True) * 100
plt.figure(figsize=(7,7))
plt.pie(proportion_album_types, labels=proportion_album_types.index, autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette('PuRd'), pctdistance=0.85)
# Add a circle at the center to make it a Donut Chart (better for diversity visualization)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Proportion of Album Types')
plt.show()

# 4. Genres Distribution (Category Diversity)
plt.figure(figsize=(12,6))
top_genres = df1['artist_genres'].value_counts().head(10)
sns.barplot(x=top_genres.index, y=top_genres.values, palette='magma')
plt.title('Top 10 Music Genres (Frequency)')
plt.xlabel('Genre')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()



```

```python
# correlation heatmap
pink = sns.light_palette("#7F025170", as_cmap=True)
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation
corr = df[['track_popularity', 'artist_popularity', 'artist_followers', 'track_duration_min', 'album_total_tracks']].corr()
sns.heatmap(corr, annot=True, cmap= pink, fmt=".2f")
plt.title('Correlation Between Music Metrics')
plt.savefig('correlation_matrix.png')
```

<h3>Trends Within The Dataset</h3> 

1. Track popularity distribution: Shows that most songs have a popularity score of 60 and 80 with a significant spike of songs at 0

2. Release trends: Displays an exponential growth in the number of songs released per year peaking in 2024/2025. this reflects the modern 'streaming era' where music is produced and uploaded more frequently than in the 20th century 

3. Top artists char: Taylor swift dominates both the number of songs and total followers and other top artist include Ed Sheeran, Billie Eilish

4. Boxplot: popularity: Explicit vs non-explicit tracks' compares the popularity scores of songs based on whether they contain explicit content. explicit tracks are significantly less popular (explicit: unsuitable for children, with swear words, mature themes, sexual content)

5. Correlation heatmap: The strongest relationship is between Artist popularity ans Artist follower(0.64). interestingly track duration and the total number of tracks on an album have no impact on popularity
